```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
study_hours = np.array([2, 4, 6, 8, 1, 5, 7, 3, 9, 10])
exam_scores = np.array([50, 60, 70, 85, 40, 65, 78, 55, 90, 95])
sleep_hours = np.array([8, 7, 6, 5, 9, 6.5, 5.5, 7.5, 4, 3])
```

```python
def pearson_correlation(x, y):
return np.corrcoef(x, y)[0, 1]
```

```python
cor_study_exam = pearson_correlation(study_hours, exam_scores)
cor_sleep_exam = pearson_correlation(sleep_hours, exam_scores)
cor_study_sleep = pearson_correlation(study_hours, sleep_hours)
```

```python
plt.figure(figsize=(15, 5))
```

```python
plt.subplot(1, 3, 1)
plt.scatter(study_hours, exam_scores, color='blue')
plt.title(f"Study Hours vs. Exam Scores (r={cor_study_exam:.2f})")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
```

```python
plt.subplot(1, 3, 2)
plt.scatter(sleep_hours, exam_scores, color='green')
plt.title(f"Sleep Hours vs. Exam Scores (r={cor_sleep_exam:.2f})")
plt.xlabel("Sleep Hours")
plt.ylabel("Exam Scores")
```

```python
plt.subplot(1, 3, 3)
plt.scatter(study_hours, sleep_hours, color='red')
plt.title(f"Study Hours vs. Sleep Hours (r={cor_study_sleep:.2f})")
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
```

```python
plt.tight_layout()
plt.show()
```
