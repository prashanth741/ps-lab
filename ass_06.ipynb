Question 1:
Generate a normal distribution of data points with a mean of 50 and a standard
deviation of 5. Plot a histogram for the generated data. Analyze the symmetry and
bell-shaped curve of a normal distribution and visualize its Probability Density
Function (PDF).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mean = 50
std_dev = 5
num_samples = 1000
```

```python
data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='skyblue', bins=30, stat='density', linewidth=1)
```

```python
plt.title('Histogram and Probability Density Function (PDF)\nfor Normal Distribution', fontsize=14)
plt.xlabel('Data Points', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()
```

Question 2:
In a sample space of 1000 cases, the mean of a certain test is 14 and standard deviation
is 2.5. Assuming distribution to be normal, Calculates and visualize the following
probabilities.
I. how many students scored between 14 and 2.5?
II. How many scored below 12?
III. How many scored above 19?


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
mean = 14
std_dev = 2.5
sample_size = 1000
```

```python
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = stats.norm.pdf(x, mean, std_dev)
```

```python
z1, z2 = (2.5 - mean) / std_dev, (14 - mean) / std_dev
p1 = stats.norm.cdf(z2) - stats.norm.cdf(z1)
students_between_2_5_and_14 = int(p1 * sample_size)
```

```python
z3 = (12 - mean) / std_dev
p2 = stats.norm.cdf(z3)
students_below_12 = int(p2 * sample_size)
```

```python
z4 = (19 - mean) / std_dev
p3 = 1 - stats.norm.cdf(z4) # Changed 'norm' to 'stats.norm' to reference the object from scipy.stats
students_above_19 = int(p3 * sample_size)
```

```python
print(f"I. Students between 2.5 and 14: {students_between_2_5_and_14}")
print(f"II. Students below 12: {students_below_12}")
print(f"III. Students above 19: {students_above_19}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, pdf, label='Normal Distribution', color='black')
```

```python
x_fill_1 = np.linspace(2.5, 14, 1000)
plt.fill_between(x_fill_1, stats.norm.pdf(x_fill_1, mean, std_dev), color='skyblue', alpha=0.5, label='Between 2.5 and 14')
```

```python
x_fill_2 = np.linspace(mean - 4*std_dev, 12, 1000)
plt.fill_between(x_fill_2, stats.norm.pdf(x_fill_2, mean, std_dev), color='orange', alpha=0.5, label='Below 12')
```

```python
x_fill_3 = np.linspace(19, mean + 4*std_dev, 1000)
plt.fill_between(x_fill_3, stats.norm.pdf(x_fill_3, mean, std_dev), color='green', alpha=0.5, label='Above 19')
```

```python
plt.title('Normal Distribution Curve with Marked Probabilities')
plt.xlabel('Test Scores')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
